This notebook explores transaction data to understand patterns
that help identify fraudulent activities.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/transactions.csv")
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x="fraud", data=df)
plt.title("Fraud vs Non-Fraud Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["transaction_amount"], bins=50)
plt.title("Transaction Amount Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
```

## Key Observations

- Fraudulent transactions are rare (high class imbalance)
- High transaction amounts correlate strongly with fraud
- International transactions increase fraud probability
- Transaction time has noticeable impact
